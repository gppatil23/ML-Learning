### MAgic Methods
Magic Methods in pythons also known as double underscore method are special methods starts & ends with double underscore(__). These methods enable you ti define the behaviour of objects for built in operations, such as arthmatic operations, comparisons or more 

Magic Method are predefined method in python that you can override to change the behaviour of your objects.

```python
'''
__init__ : Initializes a new instance of class
__str__ : returns string representation of objects 
__repr__ : Returns a official string representation of object 
__len__ : Returns len of an object
__getitem__ : Gets an item for container
__setitem__ : Sets an item for container
'''
```

```python
#Basic Methods
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name},{self.age} Years old"

    def __repr__(self):
        return f"Person(name={self.name}, age={self.age})"

person = Person("Gaurav",21)
# print(person.name)  #this is default way to call but we use __str__ method so we can return like following
print(person)  #Usage of __str__
print(repr(person))
```

### Operator Overloading

```python
class Vector:
    def __init__(self,x,y):
        self.x =x
        self.y =y
        
    def __add__(self,other):
        return self.x+other.x,self.y+other.y
    
        
v1 = Vector(3,6)
v2 = Vector(5,7)
print(v1+v2)
```

### Custom Exception(Raise and Throw Custom Exception)

```python
class Error(Exception):  #Exeption is pre-built class for raising exception
    pass

class doException(Error):
    pass
    
```

```python
year =int(input("Enter DOB"))
age = 2025 - year
#writing exception
try:
    if age<=30 and age>=20:
        print("The age is valid so you can apply")
    else:
        raise doException
except:
    print("Your age should be greater than 20 or less than 30")
```
