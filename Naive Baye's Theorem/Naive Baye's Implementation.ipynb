### Naive Baye's Theorem
#### 3 types of Naive Baye's
##### 1.Bernouli's Naive Bayes
Whenever features have only two outcomes - true/false, yes/no, pass/fail

##### 2. Multinomial Naive Bayes
whenever fratures are in textual format - column email txt sample in spam/not spam dataset
this textual information is converted into numerical form with NLP

##### 3. Gaussian Naive Bayes
features should be in continous values [graph looks like mountain in linear chart]

## Gaussian Naive Bayes Implementation(remaining 2 in NLP Later)
https://scikit-learn.org/stable/modules/generated/sklearn.naive_bayes.GaussianNB.html

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
X,y = load_iris(return_X_y=True)  #return_X_y gives dependent and independent features ##IMP
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
#importing Gaussian Naive Bayes
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
gnb.fit(X_train,y_train)
y_pred = gnb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```
