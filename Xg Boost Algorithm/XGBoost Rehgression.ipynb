## Car Price Prediction

## 1) Problem statement.
This dataset comprises used cars sold on cardehko.com in India as well as important features
of these cars.
If user can predict the price of the car based on input features.
Prediction results can be used to give new seller the price suggestion based on market
condition.

## 2) Data Collection.
The Dataset is collected from scrapping from cardheko webiste
The data consists of 13 column and 15411 rows.

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('cardekho_imputated.csv',index_col=[0])
df
```

```python
df.drop('car_name', axis=1, inplace=True)
df.drop('brand', axis=1, inplace=True)
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
X= df.drop('selling_price',axis=1)
y = df['selling_price']
```

```python
X.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X['model'] = le.fit_transform(X['model']) 
```

```python
X.head()
```

```python
num_feature =X.select_dtypes(exclude='object').columns
cat_feature = ['seller_type','fuel_type','transmission_type']

from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import ColumnTransformer

cat_encode = OneHotEncoder()
num_encode= StandardScaler()

preprocessor = ColumnTransformer([
    ('OneHotEncoder',cat_encode,cat_feature),
    ('StandardScalar',num_encode,num_feature)
],remainder='passthrough')
```

```python
X = preprocessor.fit_transform(X)
```

```python
pd.DataFrame(X)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.30,random_state=42)
```

## Model Training and Selection

```python
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,root_mean_squared_error,mean_absolute_error,mean_squared_error
```

```python
def evaluate_model(true,predicted):
    r2 = r2_score(true,predicted)
    mae = mean_absolute_error(true,predicted)
    mse = mean_squared_error(true,predicted)
    rmse = root_mean_squared_error(true,predicted)
    return r2,mae,mse,rmse
```

```python
model = XGBRegressor()
model.fit(X_train,y_train)

y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

model_train_r2,model_train_mae,model_train_mse,model_train_rmse = evaluate_model(y_train,y_train_pred)
model_test_r2,model_test_mae,model_test_mse,model_test_rmse = evaluate_model(y_test,y_test_pred)

print('model performance for training set')
print(f"Root Mean Squared error {model_train_rmse:.4f}")
print(f"Mean Squared error {model_train_mse:.4f}")
print(f"Mean Absolute error {model_train_mae:.4f}")
print(f"R2 Square {model_train_r2:.4f}")

print('='*35)

print('model performance for testing set')
print(f"Root Mean Squared error {model_test_rmse:.4f}")
print(f"Mean Squared error {model_test_mse:.4f}")
print(f"Mean Absolute error {model_test_mae:.4f}")
print(f"R2 Square {model_test_r2:.4f}")

print('\n')
```

## Hyperparameter Tunning

```python
xgboost_params = {"learning_rate": [0.1, 0.01,0.10,0.001],
                "max_depth": [5, 8, 12, 20, 30],
                "n_estimators": [100, 200, 300],
                "colsample_bytree": [0.5, 0.8, 1, 0.3, 0.4]}
```

```python
from sklearn.model_selection import RandomizedSearchCV

random = RandomizedSearchCV(estimator=XGBRegressor(),param_distributions=xgboost_params,n_iter=100,cv=3,n_jobs=-1)

random.fit(X_train,y_train)
print(random.best_params_)
```

```python
model = XGBRegressor(n_estimators=300, max_depth=5, learning_rate=0.1, colsample_bytree=0.4)
model.fit(X_train,y_train)

y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

model_train_r2,model_train_mae,model_train_mse,model_train_rmse = evaluate_model(y_train,y_train_pred)
model_test_r2,model_test_mae,model_test_mse,model_test_rmse = evaluate_model(y_test,y_test_pred)

print('model performance for training set')
print(f"Root Mean Squared error {model_train_rmse:.4f}")
print(f"Mean Squared error {model_train_mse:.4f}")
print(f"Mean Absolute error {model_train_mae:.4f}")
print(f"R2 Square {model_train_r2:.4f}")

print('='*35)

print('model performance for testing set')
print(f"Root Mean Squared error {model_test_rmse:.4f}")
print(f"Mean Squared error {model_test_mse:.4f}")
print(f"Mean Absolute error {model_test_mae:.4f}")
print(f"R2 Square {model_test_r2:.4f}")

print('\n')
```
