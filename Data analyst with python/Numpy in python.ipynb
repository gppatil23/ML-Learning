### Numpy in python
Numpy is funadamental library for scientific computing in python.It provides support for arrays and matrices, along with collection of functions to operate on these data structures. 

```python
import numpy as np
```

```python
#create an array in np
#creating 1D array
arr = np.array([1,2,4,5,6])
print(arr)
print(type(arr))

#creating 2D array
arr1 = np.array([[2,4,5],[2,4,6],[1,67,8]])
print(arr1.shape)
```

```python
#arange function - it allows you to create list having start,end,step parameter
#syntax - np.arange(start,end,step)
np.arange(0,10,2)
```

```python
#creating unit matrix (all 1's)
np.ones((3,2))
```

```python
#creating identity matrix
np.eye(3)
```

```python
#attributes of Numpy
arr = np.array([[1,2,3,5],[34,56,21,52]])

print("Array:",arr)
print("Array shape:",arr.shape)
print("Array dimention:",arr.ndim)
print("Array size(num of elements):",arr.size)
print("Array Data type:",arr.dtype)
print("Array Item size(in bytes):",arr.itemsize)

```

```python
# Numpy Vectorize Operations
arr1 = np.array([1,2,3,4,5])
arr2 = np.array([3,22,1,3,4])

## Element wise addition
print("add",arr1 +arr2)

## Element wise subtraction
print("Sub",arr1-arr2)

## Element wise Multiplication
print("Mult",arr1*arr2)

## Element wise Division
print("Div",arr1/arr2)
```

```python
#universal function
arr = np.array([1,2,3,4,5])

#square root
print(np.sqrt(arr))

#exponential
print(np.exp(arr))

#sin
print(np.sin(arr))

#natural log
print(np.log(arr))
```

```python
# array slicing and indexing operation
arr = np.array([[1,3,5,65],[2,3,15,6],[2,55,76,7]])
arr
```

```python
#accessing 15,6,76,7
#formula - arr[row_start:row_end, column_index] or arr[row_start:row_end, column_index:col_index]
#-ve column index starts from last
arr[2:,:-1]
```

```python
#modifying array
arr[0][0] = 100
print(arr)
```

```python
arr[1:] = 500
print(arr)
```

```python
#statastical concept--Normalization
#def --->to have mean 0 standard daviation of 1
data = np.array([1,2,3,4,5])
mean = np.mean(data)
std = np.std(data)

#normalize the data
noramalize_data = (data-mean) / std
print(f"Normalized data:{noramalize_data}")
```

```python
#logical operation
data1 = np.array([1,3,24,54,5,6])
# data1>5  #o/p = array([False, False, False,  True, False,  True])
data1[(data1>5) & (data1<50)]
```
