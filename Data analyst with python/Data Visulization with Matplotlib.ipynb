### DATA VISULIZATION
Matplotlib is powerful plotting library of python that enable the creation of static animated and interactive visulization. it is widely used in data visulization and data science and analytics. 

```python
import matplotlib.pyplot as plt
```

```python
x =[1,2,4,5,6]
y = [3,4,5,6,7]

#creating simple plot
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

```python
x =[1,2,4,5,6]
y = [3,4,5,6,7]
plt.plot(x,y,color ='red',linestyle = ':',marker = '+',linewidth = 3)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid(True)
```

```python
## Subplotting
x = [1,2,3,4,5,8]
y1 = [2,3,4,5,6,7]
y2 = [6,7,8,9,10,11]
plt.figure(figsize=(9,5))

plt.subplot(2,2,1) #subplot creates space for graph (row, column, ploting block number)
plt.plot(x,y1,color= 'red')
plt.title('Plot 1')

plt.subplot(2,2,2) #plots in second block
plt.plot(x,y2,color= 'blue')
plt.title('Plot 2')

# plt.subplot(1,2,3) #this won't work as it needs 2nd row to create plot 3

plt.subplot(2,2,3) #creates 3rd plotting space in 2nd row
plt.plot(y1,x,color= 'green')
plt.title('Plot 3')

plt.subplot(2,2,4) #creates 4rd plotting space in 2nd row
plt.plot(y2,x,color= 'orange')
plt.title('Plot 4')
```
