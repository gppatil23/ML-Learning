### Pandas-Dataframe and series
Pandas is powerful data manupulation library in python. widely used for data analysis and data cleaning.It provides two primary data structures. Series and Dataframe. A Series is a one-dimensional array-like object, while dataframe is 2D size, mutable, and potentially heterogeneous tabular data structure with labelled axes(rows and col)

```python
import pandas as pd
```

```python
data= [1,2,3,4,5]
series = pd.Series(data)
print(series)
```

```python
#create series from dictionary
data1 = {'a':1,'b':2,'c':3,'d':4,'e':5}
series1 = pd.Series(data1)
print(series1)
```

```python
#giving index individually
data = [1,2,3,4]
index1 = ['c','d','e','f']
pd.Series(data,index=index1)
```

```python
## Dataframe
## Create a Dataframe from dictionary of list
data = {
    'Name':['Gaurav','Vedant','Vallabh'],
    'Age':[21,26,27],
    'city':['Pune','Pune','Mumbai']
}
df = pd.DataFrame(data)
df
```

```python
df['Name']
```

```python
#accessing via index number
df.loc[0]
```

```python
df
```

```python
#accessing specialized element
df.at[1,'Name']
```

```python
# accessing via only index
df.iat[2,2]
```
