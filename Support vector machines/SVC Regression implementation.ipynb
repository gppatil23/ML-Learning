## Always remember do train test split before encoding to prevent DATA LEAKAGE

```python
df.columns
```

```python
#independent dependent features 
X = df[['tip', 'sex', 'smoker', 'day', 'time', 'size']]
y = df['total_bill']
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
X_train.head()
```

```python
#feature encoding for categorial data
# Label encoding
from sklearn.preprocessing import LabelEncoder
le1 = LabelEncoder()
le2 = LabelEncoder()
le3 = LabelEncoder()
```

```python
X_train['sex'] = le1.fit_transform(X_train['sex'])
X_train['smoker'] = le2.fit_transform(X_train['smoker'])
X_train['time'] = le3.fit_transform(X_train['time'])

#to remove warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
X_train.head()
```

```python
#same for X_test data

X_test['sex'] = le1.transform(X_test['sex'])
X_test['smoker'] = le2.transform(X_test['smoker'])
X_test['time'] = le3.transform(X_test['time'])
```

```python
X_test.head()
```

```python
# One Hot Encoding-----Column transformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
ct = ColumnTransformer(transformers=[('onehot',OneHotEncoder(drop='first'),[3])],remainder='passthrough')

#here 3 is column index of day column in form of list means 2D because it throw error not accept 1D data. remainder is set passthrough because we dont want to drop other column
#it kept encoded one and drop all when set to 'drop'
```

```python
X_train = ct.fit_transform(X_train)
```

```python
X_test = ct.transform(X_test)
```

```python
X_test  #first 3 are result of one hot encoding remaing are columns in df table
```

###  Support Vector Regression

```python
from sklearn.svm import SVR
```

```python
svr = SVR()
svr.fit(X_train,y_train)
```

```python
y_predict = svr.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
r2 = print(r2_score(y_test,y_predict))
mse = print(mean_absolute_error(y_test,y_predict))
```

### Hyper parameter Tunning

```python
from sklearn.model_selection import GridSearchCV
```

```python
para_grid = {'C':[0.1,1,10,100,1000],
            'gamma':[1,0.1,0.01,0.001],
            'kernel':['rbf']}

grid = GridSearchCV(SVR(),param_grid=para_grid,refit=True,verbose=3)
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid_pred = grid.predict(X_test)
```

```python
r2sc = print(r2_score(y_test,grid_pred))
mse_sc= print(mean_absolute_error(y_test,grid_pred))
```
